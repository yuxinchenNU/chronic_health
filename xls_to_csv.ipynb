```python
import pandas as pd
```

```python
# obtain file names
years = list(range(2010, 2019))
file_directory = '/Users/yuxinchen/Documents/Research/data_projects/data/health_ranks_data'
state = 'illinois'
filenames = []
csvfilenames = []
for year in years:
    filenames.append(file_directory + '/'  + str(year) + 'Illinois' + '.xls')
    csvfilenames.append(file_directory + '/' + state + str(year) + '.csv')
```

```python
for ind, filename in enumerate(filenames):
    df1 = pd.read_excel(filename, sheet_name = 3, skiprows=1)
    df2 = pd.read_excel(filename, sheet_name = 4, skiprows=1)
    df = pd.concat([df1, df2], axis=1, join='outer')
    df = df.iloc[:,~df.columns.duplicated()]
    df = df.drop(['Z-Score'], axis =1)
    print(df.shape[1])
    df.to_csv(csvfilenames[ind], sep=',')
```

```python
import glob
filenames = '../data/health_ranks_data/*County Health Ranking*.xls'
files = sorted(glob.glob(filenames, recursive=True))
```

```python
csvfilenames = []
for file in files:
    time = file.split(' ')[0]
    time = time.split('/')[-1]
    state = file.split(' ')[4].lower()
    csvfile = '../data/health_ranks_data/csv_files/' + state + time + '.csv'
    # read in excel files
    df1 = pd.read_excel(file, sheet_name = 3, skiprows=1)
    df2 = pd.read_excel(file, sheet_name = 4, skiprows=1)
    df = pd.concat([df1, df2], axis=1, join='outer')
    df = df.iloc[:,~df.columns.duplicated()]
    # save as cvs files
    df.to_csv(csvfile, sep=',')
    
```
